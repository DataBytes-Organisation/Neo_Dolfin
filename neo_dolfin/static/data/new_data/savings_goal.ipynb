```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('user2.csv')

savings_goal = 10000

data.fillna('', inplace=True)

data['postDate'] = pd.to_datetime(data['postDate'])
data = data.sort_values(by='postDate')
```

```python
def minimum_monthly_balance(data, date_column, balance_column):
    """
    Calculate the minimum monthly balance for given dataframe, date and balance column.
    """
    data['year_month'] = data[date_column].dt.to_period('M')

    monthly_minimum = data.groupby('year_month')[balance_column].min().reset_index()
    
    return monthly_minimum


minimum_monthly_balance(data, 'postDate', 'balance')
```
